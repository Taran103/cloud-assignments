```python
# Q1 You are a data analyst at an e-commerce company. Your task is to analyze sales data,
# clean and process it, perform statistical analysis, and generate visual insights to help the
# business make informed decisions.

import numpy as np
import pandas as pd

roll_number = 123  

np.random.seed(roll_number)
```

```python
# Generate random sales data (12 months Г— 4 categories)
sales_data = np.random.randint(1000, 5000, size=(12, 4))

sales_data
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

df = pd.DataFrame(sales_data, columns=columns, index=months)

df
```

```python
df.head()
```

```python
df.describe()
```

```python
# Total sales per category
category_totals = df.sum()
category_totals
```

```python
# Total sales per month
monthly_totals = df.sum(axis=1)
monthly_totals
```

```python
growth = df.pct_change()
avg_growth = growth.mean()

avg_growth
```

```python
# Total Sales column
df["Total Sales"] = df.sum(axis=1)

# Growth Rate column
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

df
```

```python
if roll_number % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.90
else:
    df["Clothing"] = df["Clothing"] * 0.85

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
for col in columns:
    plt.plot(df.index, df[col], label=col)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()
```

```python
# Q2 NumPy Statistical Operations

array = np.array([[1, -2, 3],
                  [-4, 5, -6]])
```

```python
np.abs(array)
```

```python
# Flattened
np.percentile(array.flatten(), [25, 50, 75])
```

```python
# Column-wise
np.percentile(array, [25, 50, 75], axis=0)
```

```python
# Row-wise
np.percentile(array, [25, 50, 75], axis=1)
```

```python
# Flattened
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std:", np.std(array))
```

```python
# Column-wise
np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0)
```

```python
# Row-wise
np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1)
```

```python
#  Q3 Rounding Functions

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Truncate:", np.trunc(a))
print("Round:", np.round(a))
```

```python
# Q4 Swap Elements in List (Using Temporary Variable)

lst = [10, 20, 30, 40]

i, j = 1, 3  # indices to swap

temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

lst
```

```python
# Q5: Swap Elements in Set (via List Conversion)

s = {10, 20, 30, 40}

# Convert to list
lst = list(s)

# Swap elements
lst[0], lst[1] = lst[1], lst[0]

# Convert back to set
s = set(lst)

s
```
